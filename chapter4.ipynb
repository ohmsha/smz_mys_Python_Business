##4.2.3　面倒な繰り返し処理はPythonに任せよう～forループ～

```python
a = [5, 2, 5, 1, 4]
```

#### コード4.6


```python
if a[1] % 2 == 0:
    print("偶数")
else:
    print("奇数")
```

#### コード4.7


```python
if a[0] % 2 == 0:
    print("偶数")
else:
    print("奇数")
if a[1] % 2 == 0:
    print("偶数")
else:
    print("奇数")
if a[2] % 2 == 0:
    print("偶数")
else:
    print("奇数")
if a[3] % 2 == 0:
    print("偶数")
else:
    print("奇数")
if a[4] % 2 == 0:
    print("偶数")
else:
    print("奇数")
```

#### コード4.8


```python
for number in a:
    if number % 2 == 0:
        print("偶数")
    else:
        print("奇数")
```

##4.2.4　独自の処理を実行しよう～関数～

#### コード4.10


```python
x = 5
y = 10
z = 2
result = (x + y) * z
result = result ** 2
result = result / 3
result = result + 10
```

#### コード4.11


```python
x = 5
y = 10
z = 2
result = (x + y) * z
result = result ** 2
result = result / 3
result = result + 10
```

#### コード4.12


```python
def complex_calculation(x, y, z):
    result = (x + y) * z
    result = result ** 2
    result = result / 3
    result = result + 10
    return result
```

#### コード4.13


```python
x = 5
y = 10
z = 2
result = complex_calculation(x, y, z)
print(result)
```

#### コード4.14


```python
x = 5
y = 10
z = 2
complex_calculation(x, y, z)
```
