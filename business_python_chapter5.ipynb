## 5.2.3　欠損の扱い方のパターンを覚えよう

```python
# コード5.13
sm_sales.isnull().sum()
```

### 欠損値の削除
- 行削除
- 列削除

```python
# コード5.14
#「合計金額」列が欠損値である行を削除する
sm_sales.dropna(subset=['合計金額'], axis=0)
```

```python
# コード5.15
# 欠損を含む全ての行を削除する
sm_sales.dropna(axis=0)
```

```python
# コード5.16
# 欠損値を含む列をすべて削除する
sm_sales.dropna(axis=1)
```

### 欠損値の補完
- 0で補完

```python
# コード5.19
# 0で補完
sm_sales.fillna(0)
```

```python
# コード5.20
sm_sales.fillna({'合計金額': 0, '支払い方法': '現金'})
```

## 5.3　データ可視化で必須のseabornでグラフを描こう
### 5.3.1　ヒストグラム

```python
# コード5.21
! pip install japanize-matplotlib
import seaborn as sns
import matplotlib.pyplot as plt
import japanize_matplotlib

# ヒストグラムの作成
plt.figure(figsize=(8, 6))
sns.histplot(sm_sales['単価'], bins=150)
plt.xlabel('単価')
plt.show()
```

### 5.3.2　箱ひげ図

```python
# コード5.22
plt.figure(figsize=(8, 6))
sns.boxplot(data=sm_sales, x='支店', y='合計金額')
plt.show()
```

### 5.3.3　棒グラフ

```python
# コード5.23
import numpy as np
plt.figure(figsize=(8, 6))
sns.barplot(data=sm_sales, x='支店', y='合計金額', estimator=np.mean)
plt.show()
```

### 5.3.4　散布図

```python
# コード5.24
plt.figure(figsize=(8, 6))
sns.scatterplot(data=sm_sales, x='単価あたりの原価', y='単価', hue='商品カテゴリ', alpha=0.8)
plt.show()
```

### 5.3.5　折れ線グラフ

```python
# コード5.25
sm_sales['購入時間'] = pd.to_datetime(sm_sales['購入時刻'], format='%H:%M').dt.hour

plt.figure(figsize=(8, 6))
sns.lineplot(data=sm_sales, x='購入時間', y='合計金額', estimator=np.mean, ci='sd')
plt.show()
```
